## 5. Errors Should Never Pass Silently

Errors are not a sign that you should stop. In Python, errors are often one of the fastest ways to learn what the language expects.

```python
# This will raise an error -- that's okay!
int("twenty")
```

Read the **last line** of the traceback first. It usually tells you the type of error and the direct cause.

In this case, Python is telling us that `"twenty"` cannot be interpreted as an integer.

```python
# The fix: use a value Python can convert
value = "20"
price = int(value)
print(price)
```

## 6. There Should Be One Obvious Way to Do It

Python often has a preferred style for common tasks. Learning these patterns helps your code become easier to write and easier to read.

```python
names = ["Alice", "Bob", "Charlie"]

# enumerate gives you both the index and the value
for index, name in enumerate(names):
    print(index, name)
```

```python
scores = [80, 90, 75]

# zip pairs up two lists element by element
for name, score in zip(names, scores):
    print(name, score)
```

## 7. Now Is Better Than Never

Python lowers the barrier to experimentation. You do not need to understand everything before you start building small things.

```python
print("Hello, ML world")
```

```python
records = [
    {"name": "A", "score": 82},
    {"name": "B", "score": 67},
    {"name": "C", "score": 91},
]

# A list comprehension filters data in one readable line
high_scores = []
for record in records:
    if record["score"] >= 80:
        high_scores.append(record)
print(high_scores)

# A list comprehension filters data in one readable line
high_scores2 = [record for record in records if record["score"] >= 80]
print(high_scores2)
```

### Reflection

> At this point, you are already doing something useful: selecting data based on a condition. That is a real step toward data analysis and ML workflows.
