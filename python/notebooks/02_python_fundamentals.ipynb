## Section 2: Core Data Types Through Chat Examples

Let's look at how chat data is represented in Python using strings, lists, and dictionaries. We'll load a real record from our sample data.

```python
# Loading a real record
sample_file = Path("sample_data/CHECK_PAYMENT_STATUS.json")
with open(sample_file, "r", encoding="utf-8") as f:
    all_records = json.load(f)

# Let's take the second record which has session history
record = all_records[1]
print("Record loaded successfully.")
```

A **String** represents text. Our `user_message` is a string.

```python
user_message = record["user_message"]

print(f"Message: {user_message}")
print(f"Type: {type(user_message)}")
print(f"Length: {len(user_message)} characters")
```

A **List** is an ordered collection. `session_history` is a list of previous conversation turns.

```python
history = record["session_history"]

print(f"History is a: {type(history)}")
print(f"Number of turns: {len(history)}")
print(f"First turn: {history[0]}")
```

A **Dictionary** stores key-value pairs. Each turn in the history is a dictionary.

```python
first_turn = history[0]

print(f"Turn type: {type(first_turn)}")
print(f"Role: {first_turn['role']}")
print(f"Content: {first_turn['content']}")
```

## Section 3: Control Flow for Data Preparation

Control flow allows us to process batches of data and handle different conditions. We'll see how to iterate through records and filter them.

We use a `for` loop to look at every record in a list.

```python
print("First 3 messages in the file:")
for rec in all_records[:3]:
    print(f"- {rec['user_message']}")
```

An `if` statement helps us handle missing data or filter records. Let's count how many records have session history.

```python
records_with_history = 0

for rec in all_records:
    if len(rec["session_history"]) > 0:
        records_with_history = records_with_history + 1

print(f"Found {records_with_history} records with session history out of {len(all_records)}.")
```

### Exercise: Filter Invalid Records
Loop through `all_records` and create a new list called `valid_records` that contains only records where `user_message` is not empty.

```python
valid_records = []
for rec in all_records:
    if rec["user_message"] != "":
        valid_records.append(rec)

print(f"Kept {len(valid_records)} valid records.")
```

## Section 4: Functions as Reusable Processing Steps

Functions help us group code that we want to run multiple times. This makes our data pipeline cleaner.

Let's create a function to calculate **session depth** (number of turns in history).

```python
def extract_session_depth(record):
    """Returns the number of turns in the session history."""
    history = record["session_history"]
    return len(history)

# Test the function
depth = extract_session_depth(record)
print(f"Session depth: {depth}")
```

## Section 5: JSON and Basic Debugging

We've been using `json.load()` to read data. JSON (JavaScript Object Notation) is a standard format that maps perfectly to Python's dictionaries and lists.

```python
# Loading another file to practice
billing_file = Path("sample_data/CHECK_BILLING_DUE_DATE.json")
with open(billing_file, "r", encoding="utf-8") as f:
    billing_data = json.load(f)

print(f"Loaded {len(billing_data)} billing records.")
```

Common errors in data processing include:
- `KeyError`: Accessing a key that doesn't exist (e.g., `record['text']` instead of `record['user_message']`)
- `TypeError`: Performing an operation on the wrong type (e.g., trying to get `len()` of an integer)

Always read the bottom of the error message to understand what went wrong!
